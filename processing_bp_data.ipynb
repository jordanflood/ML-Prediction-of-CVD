```python
import pandas as pd

# Load existing dataset
df = pd.read_csv("cleaned_cardiovascular_data.csv")

# Create new glucose categories
df['gluc_1'] = (df['gluc'] == 1).astype(int)
df['gluc_2'] = (df['gluc'] == 2).astype(int)
df['gluc_3'] = (df['gluc'] == 3).astype(int)

df['chol_1'] = (df['cholesterol'] == 1).astype(int)
df['chol_2'] = (df['cholesterol'] == 2).astype(int)
df['chol_3'] = (df['cholesterol'] == 3).astype(int)

# Combine blood pressure values
labels = [0, 1, 2]  # 0:normal, 1:elevated bp, 2:hypertension
df['bp'] = pd.cut(df['ap_hi'], [0, 120, 130, 250], right=False, labels=labels)
df.head(10)

# Drop original cholesterol, glucose, and ap columns
df = df.drop(columns=['cholesterol', 'gluc', 'ap_hi', 'ap_lo'])

# Save updated DataFrame to a new CSV
df.to_csv("final_cardiovascular_data.csv", index=False)
```
